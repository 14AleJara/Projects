```python
df2 = df
```

```python
df2b = df2
```

```python
df2 = df2.drop(columns = ['Online Interface - Ease of Use',
                    'Online Interface - Ease of Access',
                    'Online Interface - Navigation',
                    'Online Interface - Likelihood to Recommend',
                    'Online Interface - Overall Rating'])
```

```python
df2b = df2b.drop(columns = ['Mobile App - Ease of Use',
                    'Mobile App - Ease of Access',
                    'Mobile App - Navigation',
                    'Mobile App - Likelihood to Recommend',
                    'Mobile App - Overall Rating'])
```

```python
df2 = df2.melt('Customer ID', var_name='Mobile App', value_name='Value')
df2
```

```python
df2b = df2b.melt('Customer ID', var_name='Online Interface', value_name='Value')
df2b
```

```python
df2['Mobile App'] = df2['Mobile App'].str.replace('Mobile App - ',"")
```

```python
df2b['Online Interface'] = df2b['Online Interface'].str.replace('Online Interface - ',"")
```

```python
df2
```

```python
df2 = df2.rename(columns = {'Mobile App':'Category', 'Value':'Mobile App'})
```

```python
df2b = df2b.rename(columns = {'Online Interface':'Category', 'Value':'Online Interface'})
```

```python
df3 = df2.merge(df2b)
df3
```

```python
df3 = df3.groupby('Customer ID').mean()
```

```python
df3
        
```

```python
df3['Difference'] = df3['Mobile App'] - df3['Online Interface']
```

```python
df3
```

```python
#defining function filter 
def filter(x):
    if x >= 2:
        return 'Mobile App Superfan'
    if (x < 2 and x >= 1):
        return 'Mobiel App Fan'
    if (x < 1 and x > -1):
        return 'Neutral'
    if (x <= -1 and x > -2):
        return 'Online Interface Fan'
    if x <= -2:
        return 'Online Interface Superfan'
    
```

```python
df3['Fan Level'] = df3['Difference'].apply(filter)
```

```python
df3
```

```python
df4 = df3.groupby('Fan Level').count()
```

```python
df4
```

```python
df4 = df4.drop(columns = ['Mobile App', 'Online Interface'])
```

```python
df4
```

```python
df4['% Total'] = 100 * df4['Difference'] / df4['Difference'].sum()
df4
```

```python
df4['% Total'] = df4['% Total'].round(2)
```
