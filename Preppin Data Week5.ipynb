```python
import pandas as pd
```

```python
df = pd.read_csv(r'/Users/constanzavargasmier/Desktop/Ale/DATASETS/PD 2023 Wk 1 Input.csv')
```

```python
df
```

```python
df[['Bank','Transaction_Code_2']] = df['Transaction Code'].str.split('-',1, expand=True)
```

```python
df
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date']).dt.strftime('%B')
df
```

```python
df = df.drop(columns = ['Transaction_Code_2', 'Transaction Code', 'Customer Code', 'Online or In-Person'])
df
```

```python
df2 = df.groupby(['Transaction Date','Bank']).sum()
df2
```

```python
df3 = df
```

```python
df3 = df3.groupby(['Transaction Date','Bank']).sum().sort_values(by = ['Transaction Date','Value'], ascending=False)
```

```python
df3
```

```python
df3['Rank'] = 1
```

```python
df3
```

```python
df3['Rank'] = df3.groupby(['Transaction Date'])['Rank'].cumsum()
```

```python
df3
```

```python
df3.rename(columns = {'Rank':'Bank_Rank_Month'}, inplace=True)
```

```python
df3
```

```python
df4 = df3
```

```python
df4['Avg_Transaction_Per_Rank'] = df4.groupby('Bank_Rank_Month').transform('mean')
```

```python
df4['Avg_Transaction_Per_Rank'] = df4['Avg_Transaction_Per_Rank'].round(2)
```

```python
df4
```

```python
df4[['Avg_Rank_Bank','Avg_Rank_Bank_2','Avg_Rank_Bank_3']] = df4.groupby('Bank').transform('mean')
```

```python
df4
```

```python
df4.drop(columns = ['Avg_Rank_Bank','Avg_Rank_Bank_3'], inplace=True)
df4
```

```python
df4 = df4.rename(columns = {'Avg_Rank_Bank_2': 'Avg_Rank_Bank'})
```
